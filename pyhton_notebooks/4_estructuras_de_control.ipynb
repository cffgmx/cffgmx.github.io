## La instrucción `if`

La idea es directa: *si* una condición es verdadera, ejecuta este bloque de código; de lo contrario, ignóralo. En Python esa idea se expresa con la palabra clave `if`, seguida de la condición y dos puntos. El bloque de instrucciones que debe ejecutarse cuando la condición es `True` va en las líneas siguientes, desplazado hacia la derecha mediante indentación, exactamente igual que el cuerpo de una función.

```python
if condición:
    # instrucciones que se ejecutan solo si condición es True
```

La condición puede ser cualquier expresión que Python pueda evaluar como `True` o `False`: una comparación, un operador lógico, o incluso el resultado de una función. El bloque indentado se ejecuta únicamente si esa evaluación resulta `True`; si resulta `False`, Python salta directamente a la siguiente línea no indentada.

````python
temperatura = 38.5

if temperatura > 37.5:
    print("Alerta: temperatura fuera del rango normal.")
````

Si cambiamos el valor de `temperatura` a uno dentro del rango normal, el `if` no se cumple y el mensaje no aparece: el bloque indentado simplemente no se ejecuta.

````python
temperatura = 36.8

if temperatura > 37.5:
    print("Alerta: temperatura fuera del rango normal.")

print("Lectura procesada.")   # Esta línea siempre se ejecuta
````

Vale la pena detenerse en la indentación, porque genera una duda natural: ¿qué líneas pertenecen al `if` y cuáles quedan fuera? La respuesta es precisa: pertenecen al `if` todas las líneas que están desplazadas hacia la derecha respecto a él. En cuanto una línea vuelve al mismo nivel de indentación que el `if`, queda fuera del bloque y se ejecuta siempre, independientemente de si la condición fue `True` o `False`.

En el ejemplo anterior, `print("Lectura procesada.")` no está indentada, por lo que se ejecuta siempre. La indentación no es un detalle estético: es la forma en que Python determina la estructura del programa. El mismo principio que vimos con las funciones aplica aquí de forma idéntica.

## La cláusula `else`

Con solo `if`, el código puede reaccionar cuando algo es verdadero, pero no tiene nada que decir cuando es falso. Muchas situaciones exigen exactamente eso: hacer una cosa *si* se cumple la condición, y otra cosa *si no* se cumple. Para ese caso existe la cláusula `else`.

`else` se escribe al mismo nivel de indentación que el `if` al que pertenece, seguido de dos puntos. Su bloque se ejecuta únicamente cuando la condición del `if` resultó `False`; nunca ambos bloques se ejecutan al mismo tiempo.

````python
calificacion = 7.5

if calificacion >= 6.0:
    print("Aprobado")
else:
    print("Reprobado")
````

La estructura `if` / `else` garantiza que exactamente uno de los dos bloques se ejecutará. No importa qué valor tome `calificacion`: si es mayor o igual a `6.0` se imprime `"Aprobado"`; en cualquier otro caso, `"Reprobado"`. No existe un tercer camino.

Esta estructura también es útil cuando el resultado de la condición debe guardarse en una variable. Un patrón común es asignar un valor distinto dependiendo de si algo se cumple o no:

````python
precio_base = 250.0
es_socio    = True

if es_socio:
    precio_final = precio_base * 0.85   # 15 % de descuento para socios
else:
    precio_final = precio_base

print(f"Precio final: ${precio_final:.2f}")
````

Aquí la condición es directamente la variable `es_socio`, cuyo valor ya es `True`. Python la evalúa tal cual: no es necesario escribir `es_socio == True`. La variable booleana ya *es* la condición. Del mismo modo, si en algún momento se quisiera comprobar que alguien *no* es socio, la forma natural es `not es_socio`, en lugar de escribir `es_socio == False`. Estas formas más directas no son solo una cuestión de estilo: son las que los programadores esperan leer y producen código más claro.

## La cláusula `elif`

En ocasiones las posibilidades no se reducen a dos: verdadero o falso, aprobado o reprobado. Una escala de calificaciones tiene varios niveles; una clasificación de riesgo puede ser baja, media o alta; una tarifa puede depender de tres o más tramos. Para manejar más de dos casos de forma limpia y ordenada, Python ofrece la cláusula `elif`, contracción de *else if*.

`elif` permite encadenar condiciones adicionales entre el `if` inicial y el `else` final. Python las evalúa en orden, de arriba hacia abajo, y ejecuta el bloque de la primera que resulte `True`. Una vez que encuentra esa condición, salta todas las demás y continúa después del bloque completo. Si ninguna condición resulta `True`, se ejecuta el bloque del `else`, si existe.

````python
calificacion = 8.2

if calificacion >= 9.0:
    nivel = "Sobresaliente"
elif calificacion >= 7.0:
    nivel = "Bien"
elif calificacion >= 6.0:
    nivel = "Suficiente"
else:
    nivel = "Reprobado"

print(f"Calificación: {calificacion} → {nivel}")
````

Con `calificacion = 8.2`, Python evalúa primero `calificacion >= 9.0`: es `False`. Pasa al primer `elif`: `calificacion >= 7.0` es `True`. Ejecuta ese bloque, asigna `"Bien"` a `nivel`, y salta todo lo demás. Los `elif` restantes y el `else` no se evalúan en absoluto.

Una duda que surge con frecuencia al ver este código es: ¿por qué no es necesario escribir `calificacion >= 7.0 and calificacion < 9.0` en el primer `elif`? La respuesta está en el orden de evaluación: si la ejecución llega al primer `elif`, ya se sabe que la condición del `if` fue `False`, es decir, que `calificacion < 9.0`. Python ha descartado ese caso. El `elif` puede asumir ese contexto y verificar solo la condición nueva que le corresponde.

Esto también revela por qué el orden de las condiciones importa. Si la primera condición fuera `calificacion >= 6.0` en lugar de `calificacion >= 9.0`, cualquier calificación de 6 o más caería siempre en ese primer bloque, sin importar cuán alta fuera. La regla práctica al encadenar intervalos numéricos es escribir primero el caso más restrictivo, es decir, el que exige el valor más alto.

## Condiciones compuestas

Los operadores lógicos `and`, `or` y `not` que ya conocemos cobran aquí su pleno significado. Una condición de `if` no tiene que ser una comparación simple: puede ser cualquier expresión booleana, por compleja que sea. Esto permite expresar restricciones que involucran varias variables o varios rangos en una sola condición.

Por ejemplo, para clasificar un punto del plano según el cuadrante en que se encuentra, es necesario verificar dos condiciones al mismo tiempo:

````python
x =  3.0
y = -1.5

if x > 0 and y > 0:
    cuadrante = "I"
elif x < 0 and y > 0:
    cuadrante = "II"
elif x < 0 and y < 0:
    cuadrante = "III"
elif x > 0 and y < 0:
    cuadrante = "IV"
else:
    cuadrante = "sobre un eje"

print(f"El punto ({x}, {y}) está en el cuadrante {cuadrante}.")
````

El operador `in`, que ya usamos con secuencias y conjuntos para preguntar si un elemento está presente, también funciona directamente como condición de un `if`. Esto resulta natural cuando se quiere verificar si un valor pertenece a un grupo de opciones válidas, evitando encadenar varios `or`:

````python
dia = "sábado"

if dia in {"sábado", "domingo"}:
    print(f"{dia} es fin de semana.")
else:
    print(f"{dia} es día hábil.")
````

Comparado con `dia == "sábado" or dia == "domingo"`, la versión con `in` es más concisa y más fácil de extender: si mañana hubiera que agregar un tercer caso, bastaría con añadirlo al conjunto sin modificar la estructura de la condición.

## `if` anidados

Los bloques indentados dentro de un `if` pueden contener cualquier instrucción válida de Python, incluyendo otros `if`. Un `if` dentro de otro `if` se llama *anidado*, y sirve para expresar decisiones que dependen de condiciones que solo tienen sentido verificar una vez que ya se cumplió otra.

El caso típico es cuando una segunda condición solo es relevante si la primera ya fue verdadera. Para determinar si un número es positivo y par, verificar la paridad solo tiene sentido si antes ya sabemos que el número es positivo:

````python
numero = 14

if numero > 0:
    if numero % 2 == 0:   # El operador % devuelve el residuo de la división entera
        print(f"{numero} es positivo y par.")
    else:
        print(f"{numero} es positivo e impar.")
else:
    print(f"{numero} no es positivo.")
````

El anidamiento puede crecer varios niveles, pero conviene usarlo con moderación. Cuando la lógica exige muchos niveles de indentación, el código se vuelve difícil de leer. En esos casos suele ser posible reescribir las condiciones usando `and` para aplanar la estructura:

```python
if numero > 0 and numero % 2 == 0:
    print(f"{numero} es positivo y par.")
```

La elección entre anidamiento y condición compuesta depende de cuál forma comunica mejor la intención. Si las dos condiciones están conceptualmente al mismo nivel, la versión con `and` suele ser más clara. Si la segunda condición es un refinamiento que solo tiene sentido dentro del contexto de la primera, el anidamiento puede comunicar esa dependencia con mayor fidelidad.

## `if` dentro de funciones y el `return` anticipado

En el notebook sobre funciones vimos que `return` entrega el resultado de una función al código que la llamó, y que ese es el único valor que cruza la frontera entre el interior y el exterior de la función. Lo que no mencionamos entonces —porque aún no teníamos las herramientas para verlo— es que una función puede tener más de un `return`, y que en cuanto Python ejecuta cualquiera de ellos, la función termina de inmediato. Las líneas que vengan después simplemente no se ejecutan.

Esto se vuelve muy natural al combinar funciones con estructuras `if`. Una función que clasifica una calificación puede devolver su resultado directamente desde cada rama, sin necesidad de guardar el valor en una variable intermedia:

````python
def clasificar(calificacion):
    if calificacion >= 9.0:
        return "Sobresaliente"
    elif calificacion >= 7.0:
        return "Bien"
    elif calificacion >= 6.0:
        return "Suficiente"
    else:
        return "Reprobado"

print(clasificar(8.2))
print(clasificar(5.9))
````

Cada rama del `if` / `elif` / `else` tiene su propio `return`. Cuando Python encuentra el primero que aplica, entrega ese valor y la función concluye. Los demás `return` ni siquiera se consideran. El resultado es exactamente el mismo que con una variable intermedia, pero el código es más directo: no hay que leer hasta el final para saber qué devuelve cada caso.

Una pregunta que surge al ver múltiples `return` en una función es si esto es una buena práctica, dado que en matemáticas una función tiene exactamente una salida. La respuesta es que en programación, los múltiples `return` no contradicen esa idea: la función sigue produciendo exactamente un resultado por cada llamada. Lo que cambia es qué resultado produce dependiendo de los datos que recibe. Desde fuera, el comportamiento es idéntico al de una función con un solo `return`.

### El `return` anticipado como mecanismo de validación

El `return` dentro de un `if` no solo sirve para entregar resultados desde distintas ramas: también es la base de un patrón de diseño muy frecuente en la práctica, conocido como *retorno temprano* o *cláusula de guarda*. La idea es sencilla: si los datos que recibe la función no son válidos para que el cálculo tenga sentido, se detiene la ejecución de inmediato —antes de hacer cualquier otra cosa— y se devuelve un valor que indique el problema.

Consideremos una función que calcula el promedio de una lista. El promedio requiere dividir por la cantidad de elementos, operación que no tiene sentido si la lista está vacía. La cláusula de guarda resuelve ese caso al inicio, sin necesidad de envolver todo el cuerpo de la función en un `if`:

````python
def promedio(valores):
    if not valores:       # True si la lista está vacía
        return None       # No tiene sentido calcular el promedio de nada
    return sum(valores) / len(valores)

print(promedio([7.5, 9.0, 8.5]))   # Lista con elementos: calcula normalmente
print(promedio([]))                # Lista vacía: devuelve None
````

Hay tres ideas en este código que merecen atención.

La primera es la condición `not valores`. Una pregunta natural es: ¿por qué funciona evaluar una lista directamente como condición de un `if`? Python considera que una colección vacía —lista `[]`, tupla `()`, cadena `""`, conjunto `set()`, diccionario `{}`— es equivalente a `False` en un contexto booleano, y cualquier colección con al menos un elemento equivale a `True`. Este comportamiento es intencional y se usa con frecuencia en código real. Por eso `not valores` es `True` exactamente cuando la lista está vacía, que es el caso que queremos detectar.

La segunda idea es `None`. Es un valor especial de Python que representa la ausencia de un resultado significativo: no es cero, no es una cadena vacía, es simplemente *nada*. Cuando una función no tiene nada útil que devolver en un caso determinado, `None` es la forma canónica de expresarlo. Una función que devuelve `None` comunica al código que la llama: *el cálculo no pudo realizarse*.

La tercera idea es la estructura que produce el retorno temprano. Sin él, habría que envolver el cálculo dentro de un `else`, y el cuerpo principal de la función quedaría un nivel más adentro:

```python
def promedio(valores):
    if not valores:
        return None
    else:                                    # este else es innecesario
        return sum(valores) / len(valores)   # un nivel extra de indentación
```

Con el retorno temprano, el `else` sobra: si la lista está vacía, la función ya terminó. Lo que viene después solo puede ejecutarse si la lista tiene elementos. El código se lee de forma más plana y la lógica principal no queda enterrada bajo condiciones de validación.

El patrón escala de forma natural cuando hay varias condiciones que verificar antes de poder calcular algo. Cada cláusula de guarda ocupa su propia línea al principio de la función, y el cuerpo principal permanece legible e indentado al primer nivel:

````python
def velocidad(distancia, tiempo):
    if distancia < 0:
        return None   # La distancia no puede ser negativa
    if tiempo <= 0:
        return None   # El tiempo debe ser estrictamente positivo
    return distancia / tiempo

print(velocidad(100, 2))    # Caso válido: 50.0
print(velocidad(100, 0))    # Tiempo cero: devuelve None
print(velocidad(-10, 2))    # Distancia negativa: devuelve None
````

Aquí aparecen dos `if` separados, no encadenados con `elif`. Esto es intencional: cada guarda verifica un problema distinto e independiente, y ambas devuelven `None`. Usar `elif` en lugar de dos `if` separados también funcionaría en este caso específico —porque en cuanto la primera guarda se cumple, la función ya terminó—, pero dos `if` independientes comunican con más claridad que se trata de dos validaciones distintas, no de ramas alternativas de una misma decisión.

El cálculo real, que ocupa la última línea, solo se alcanza cuando todos los controles anteriores han pasado. Esta organización —validar primero, calcular al final— es un hábito que hace el código más robusto y más fácil de mantener.

## Ejemplo integrador

Para terminar, conviene ver cómo las estructuras de este notebook se combinan de forma natural con las del notebook anterior. Un diccionario puede representar los datos de un producto; una función con cláusulas de guarda y ramas `if` / `elif` / `else` puede calcular su precio final según distintas reglas de negocio. El resultado es un pequeño programa que ya hace algo no trivial: valida los datos de entrada, los evalúa contra condiciones y produce una respuesta diferenciada.

````python
def calcular_precio_final(producto):
    # Cláusulas de guarda: validamos que los datos sean coherentes
    if "precio" not in producto:
        return None   # No podemos calcular nada sin precio base
    if producto["precio"] < 0:
        return None   # Un precio negativo no tiene sentido

    precio = producto["precio"]

    # Determinamos el descuento según categoría y disponibilidad en oferta
    if producto["en_oferta"]:
        if producto["categoria"] == "electrónica":
            descuento = 0.12   # 12 % para electrónica en oferta
        elif producto["categoria"] == "ropa":
            descuento = 0.20   # 20 % para ropa en oferta
        else:
            descuento = 0.08   # 8 % para otras categorías en oferta
    else:
        descuento = 0.0        # Sin descuento fuera de oferta

    return precio * (1 - descuento)


laptop = {
    "nombre":    "Laptop",
    "precio":    18500.0,
    "categoria": "electrónica",
    "en_oferta": True
}

precio_final = calcular_precio_final(laptop)

if precio_final is not None:
    print(f"Producto:     {laptop['nombre']}")
    print(f"Precio base:  ${laptop['precio']:,.2f}")
    print(f"Precio final: ${precio_final:,.2f}")
else:
    print("Error: datos del producto incompletos o inválidos.")
````

Vale la pena señalar el uso de `is not None` en la última condición. Para verificar si una variable es exactamente `None`, Python recomienda usar `is` en lugar de `==`. El operador `is` comprueba identidad: pregunta si las dos cosas son el mismo objeto en memoria, no solo si tienen el mismo valor. En el caso de `None`, que es un objeto único en Python, `is None` e `is not None` son las formas convencionales y se ven con frecuencia en código real.

También vale notar que en este ejemplo hemos accedido a las claves del diccionario con la notación de corchetes que ya conocemos: `producto["en_oferta"]`, `producto["categoria"]`. A diferencia de la versión anterior, no se usa ningún mecanismo nuevo: todo lo que aparece aquí ya fue explicado en los notebooks previos.

Las estructuras de control que hemos visto en este notebook resuelven el problema de la decisión: el código puede ahora elegir entre distintos caminos según los datos que recibe. Pero hay un segundo tipo de problema que todavía no hemos abordado: la repetición.

Imaginemos que tenemos una lista con las calificaciones de todos los alumnos de un grupo y queremos clasificar a cada uno usando exactamente la función `clasificar` que definimos en este notebook. Con las herramientas actuales, la única forma de hacerlo sería llamar a esa función una vez por alumno, escribiendo una línea distinta para cada uno. Eso es insostenible cuando el grupo tiene treinta personas, e imposible cuando el tamaño de la lista no se conoce de antemano.

Lo que necesitamos es una forma de decirle a Python: *repite esta secuencia de instrucciones para cada elemento de esta colección*. Esa capacidad —recorrer una colección y operar sobre cada uno de sus elementos— es lo que introduce el siguiente notebook, y es uno de los mecanismos más poderosos del lenguaje.
