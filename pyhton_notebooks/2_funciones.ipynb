En los notebooks anteriores aprendimos a guardar valores en variables y a agruparlos en secuencias. Pero hay algo que todavía falta: una forma de nombrar no solo datos, sino *procesos*. En matemáticas, esa idea tiene nombre propio: una *función* es una regla que toma uno o más valores de entrada y produce exactamente un valor de salida. La notación $f(x) = x^2$ describe esa regla con precisión: la función se llama $f$, recibe $x$, y devuelve su cuadrado. Lo que importa no es el resultado concreto, sino la idea general: una función es un procedimiento que siempre aplica la misma operación con los datos que recibe.

Python adopta exactamente esa idea. Una función en Python es un bloque de código con nombre al que se le pueden pasar datos y que produce un resultado. La notación también es similar: en lugar de $f(x)$, escribimos `nombre(dato)`. Los paréntesis no son decorativos; son la señal de que estamos *llamando* a la función, es decir, pidiéndole que ejecute su tarea.

Python incluye de fábrica un conjunto de funciones listas para usar, sin necesidad de definir nada. Se llaman funciones *built-in*, y cubren las operaciones más comunes. Una de ellas es `len()`, que recibe una secuencia y devuelve cuántos elementos tiene.

````python
temperaturas = [23.5, 24.0, 22.8, 19.0]
len(temperaturas)
````

````python
len("Python")
````

De manera análoga, `min()` y `max()` devuelven el menor y el mayor de los elementos de una secuencia. También aceptan varios valores separados por comas, sin necesidad de envolverlos en una lista.

````python
min(temperaturas)
````

````python
max(23.5, 24.0, 22.8)  # También acepta valores directos, sin lista
````

La función `round()` recibe un número decimal y devuelve su aproximación entera más cercana. Como veremos más adelante, también permite conservar un número específico de decimales.

````python
round(22.8673)
````

Existe una función que merece una mención especial porque se comporta de forma distinta a las anteriores: `print()`. Mientras que `len()`, `min()`, `max()` y `round()` calculan un valor y lo *devuelven* —es decir, producen un resultado que puede guardarse en una variable o usarse en una expresión—, `print()` no devuelve ningún valor útil. Su propósito es simplemente mostrar información en pantalla, hacer visible algo para la persona que está ejecutando el programa.

````python
print("Hola mundo")
````

## Funciones propias

Las funciones built-in cubren operaciones generales, pero tarde o temprano aparecen necesidades más específicas que ninguna función predefinida satisface. En esos casos, Python permite crear funciones propias. Y aquí aparece uno de los hábitos más importantes al programar: en lugar de escribir los mismos pasos una y otra vez en distintas partes del programa, se escriben una sola vez dentro de una función y se reutilizan cuantas veces sea necesario.

Para definir una función se usa la palabra clave `def`, seguida del nombre de la función, unos paréntesis y dos puntos. Dentro de los paréntesis se declaran los *parámetros*: los nombres de los valores que la función recibirá cuando se use. Por ejemplo, para convertir temperaturas de Celsius a Fahrenheit:

```python
def convertir_a_fahrenheit(celsius):
```

Aquí, `convertir_a_fahrenheit` es el nombre de la función, y `celsius` es el parámetro, es decir, el nombre que usamos dentro de la función para referirnos al valor que recibirá. Es solo una etiqueta: la función se comportaría igual si el parámetro se llamara `x`.

En la línea siguiente, desplazado hacia la derecha, aparece el *cuerpo de la función*: las instrucciones que se ejecutarán cuando se llame. Ese desplazamiento se llama *indentación*, y es la forma en que Python distingue qué líneas pertenecen a la función y cuáles quedan fuera de ella.

Dentro del cuerpo, la palabra clave `return` cumple un papel central: indica el valor que la función *entrega* al código que la llamó. Sin `return`, la función ejecuta sus instrucciones pero no produce ningún resultado aprovechable; es como hacer un cálculo en un papel que nadie puede leer. Con `return`, el resultado queda disponible: puede guardarse en una variable, usarse en una expresión o pasarse como argumento a otra función. En la analogía matemática, `return` es el signo `=` de la notación $f(x) = x^2$: especifica qué es lo que la función produce.

```python
def convertir_a_fahrenheit(celsius):
    resultado = (celsius * 9/5) + 32    # dentro de la función: indentado
    return resultado                    # el valor que la función entrega
print("esto no es parte de la función") # sin indentar: fuera de la función
```

Esta distinción entre lo que está dentro y fuera de la función tiene otra consecuencia importante: las variables que se crean dentro del cuerpo son *variables locales*, y solo existen mientras la función se está ejecutando. Una vez que `return` entrega su resultado y la función termina, esas variables desaparecen; no pueden leerse ni modificarse desde fuera. En el ejemplo anterior, `resultado` es una variable local: existe solo dentro de `convertir_a_fahrenheit` y no es accesible en el resto del programa. Esto no es una limitación accidental, sino un diseño deliberado: permite que distintas funciones usen los mismos nombres de variables sin que ninguna interfiera con las otras.

````python
def convertir_a_fahrenheit(celsius):
    resultado = (celsius * 9/5) + 32
    return resultado
# la siguiente línea ya no es parte de la función
convertir_a_fahrenheit(0) # aquí usamos nuestra función evaluada en 0
````

````python
# Esta celda produce un error: 'resultado' es local a convertir_a_fahrenheit
#resultado   # NameError: name 'resultado' is not defined
````

Si descomentamos la línea anterior, Python produce un `NameError`: el nombre `resultado` no existe fuera de la función. Eso confirma que la variable es estrictamente local. El único valor que cruza la frontera entre el interior y el exterior de la función es el que se entrega explícitamente con `return`.

Al llamar a la función se le pasan valores concretos entre paréntesis —en este caso, el número `0`—. Esos valores concretos se llaman *argumentos*, y se distinguen de los parámetros en que los parámetros son los nombres declarados en la definición, mientras que los argumentos son los valores reales con los que se ejecuta la función.

Una función puede recibir más de un parámetro, separados por comas, en directa correspondencia con la notación $f(x, y)$. Al llamarla, los argumentos se asignan a los parámetros en el mismo orden: el primero al primero, el segundo al segundo. Por esa razón se les llama *argumentos posicionales*.

````python
def dividir(dividendo, divisor):
    return dividendo / divisor

dividir(10, 2)    # aquí dividendo=10 y divisor=2
````

````python
dividir(2, 10) # aquí dividendo=2 y divisor=10
````

En esencia, crear una función es darle un nombre a un proceso para poder reutilizarlo. Este hábito —identificar las operaciones que se repiten y convertirlas en funciones— es uno de los más importantes al empezar a programar. El código se vuelve más claro, más ordenado y más fácil de modificar. Además, esta forma de pensar será la base para trabajar con estructuras más complejas que veremos más adelante.

Ahora bien, para usar una función con eficacia no basta con saber llamarla: también conviene saber qué parámetros acepta, cuáles son obligatorios y cuáles tienen un valor por defecto. Esa información está siempre disponible.

## Cómo explorar funciones y encontrar más

Las funciones que hemos visto —`len()`, `min()`, `max()`, `round()`, `print()`— son solo una muestra pequeña de las funciones built-in de Python. En realidad el lenguaje incluye muchas más, y conocerlas todas de memoria no es ni posible ni necesario. Lo importante es saber *cómo encontrar* lo que se necesita cuando se necesita.

La herramienta más inmediata está dentro del propio notebook: la función `help()`. Pasándole como argumento el nombre de otra función, Python muestra su documentación básica: qué parámetros acepta, qué valor devuelve y una breve descripción de su comportamiento.

````python
help(round)
````

La salida de `help()` puede parecer densa a primera vista, pero casi siempre sigue la misma estructura. Primero aparece la *firma de la función*: su nombre seguido de los parámetros que acepta. Después viene una descripción de lo que hace y, a veces, detalles adicionales sobre su comportamiento.

La firma es la parte más útil. Muestra qué parámetros son obligatorios y cuáles son opcionales. Cuando un parámetro aparece con un `=`, como `ndigits=None` en el caso de `round()`, significa que tiene un valor por defecto y puede omitirse: si no se proporciona, Python usa el valor que aparece después del `=`.

En el caso de `round()`, la firma nos dice que acepta dos parámetros: `number`, el número que queremos redondear, y `ndigits`, el número de decimales que queremos conservar. Como `ndigits` tiene valor por defecto `None`, podemos simplemente llamar a `round(número)` y Python redondeará al entero más cercano. Si escribimos `round(número, 2)`, conservará dos cifras decimales.

Vale la pena hacer lo mismo con `print()`, que acepta más opciones de las que hemos usado hasta ahora.

````python
help(print)
````

La firma de `print()` revela algo interesante: todos sus parámetros opcionales tienen nombre (`sep`, `end`, `file`, `flush`). Esto conecta directamente con una característica del lenguaje que ya está implícita en lo que hemos visto: Python permite pasar argumentos usando explícitamente el nombre del parámetro al que corresponden, con la sintaxis `nombre=valor`. A estos se les llaman *argumentos por nombre* o *keyword arguments*.

La ventaja es clara: con keyword arguments, el orden deja de importar, porque cada valor llega directamente al parámetro que le corresponde, sin ambigüedad.

````python
dividir(dividendo=10, divisor=2)    # explícito: no depende del orden
````

````python
dividir(divisor=2, dividendo=10)    # el orden cambió, el resultado no
````

Los keyword arguments son especialmente útiles cuando una función tiene varios parámetros opcionales y el orden no es obvio a primera vista. Con `print()`, por ejemplo, permiten modificar comportamientos específicos sin tener que recordar en qué posición va cada parámetro.

````python
print("manzana", "pera", "uva", sep=" | ")    # sep es keyword argument
````

Siguiendo con `print()`, hay algo más que conviene conocer: los *f-strings*. Son una forma de construir cadenas que mezclan texto fijo con valores de variables, todo en una sola línea. Se escriben poniendo una `f` antes de las comillas de apertura, y cualquier variable o expresión que se quiera insertar se encierra entre llaves `{}`.

````python
ciudad = "Ciudad de México"
temp = 23.5

print(f"La temperatura en {ciudad} es de {temp} grados.")
````

Para explorar más allá de lo que `help()` muestra, la fuente más completa es la documentación oficial de Python en *docs.python.org*. La sección *Built-in Functions* lista todas las funciones disponibles con ejemplos y explicaciones detalladas. Es un buen lugar al que volver cuando se necesita saber si ya existe una función que haga algo antes de intentar escribirla desde cero.

En la práctica, la forma más rápida de resolver una duda concreta no es leer la documentación de arriba a abajo, sino buscar directamente. Una búsqueda como *"python redondear número a 2 decimales"* o *"python cómo saber si un elemento está en una lista"* lleva casi siempre a resultados útiles, frecuentemente en *Stack Overflow*, un sitio donde programadores publican preguntas y respuestas técnicas.

Una alternativa cada vez más útil es consultar a un asistente de inteligencia artificial como Claude o ChatGPT. A diferencia de una búsqueda que devuelve una lista de páginas, una IA puede responder preguntas concretas en lenguaje natural, adaptar los ejemplos al contexto en que se está trabajando, o explicar un error que apareció al ejecutar una celda. Como cualquier herramienta, tiene limitaciones —a veces produce respuestas incorrectas con aparente seguridad—, así que conviene verificar lo que sugiere ejecutando el código que proponga. Pero para aprender, explorar y desbloquear problemas concretos, es un recurso valioso.

Con el tiempo la secuencia se vuelve natural: si se conoce el nombre de la función, `help()` resuelve la duda en segundos; si se necesita explorar qué existe, la documentación oficial es el lugar; si se tiene una tarea concreta pero no se sabe cómo resolverla, una búsqueda bien formulada o una pregunta a una IA es el camino más directo. Ninguna de estas herramientas excluye a las otras: los programadores las usan todas, con frecuencia en la misma sesión de trabajo.
