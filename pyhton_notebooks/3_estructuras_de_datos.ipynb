El concepto de *estructura de datos* es más amplio. Una estructura de datos es cualquier forma de organizar y almacenar información de modo que pueda usarse eficientemente. Todas las secuencias son estructuras de datos, pero no todas las estructuras de datos son secuencias: algunas colecciones no tienen orden, y otras no se recorren por posición sino por nombre. En este notebook se presentan las cuatro estructuras fundamentales de Python: la *lista*, la *tupla*, el *conjunto* y el *diccionario*.

Cada una organiza los datos de una manera distinta, y entender esas diferencias es lo que permite elegir la herramienta adecuada para cada problema. ## Listas `list`

La lista ya es conocida del notebook anterior. Conviene ahora consolidar su descripción desde una perspectiva más amplia: una *lista* es una colección *ordenada* y *mutable*. Ordenada significa que los elementos tienen una posición fija; mutable significa que ese contenido puede cambiar después de crear la lista.

La lista modela situaciones donde los datos crecen o cambian con el tiempo. Las lecturas de temperatura de un sensor a lo largo del día son el ejemplo natural: cada hora se registra un nuevo valor que se agrega a la colección. No tiene sentido usar una estructura rígida para datos que por naturaleza son dinámicos.

Se definen con corchetes `[]`, separando los elementos por comas. Por el momento solo se muestra cómo *reemplazar* un elemento existente; agregar y eliminar elementos requiere el uso de *métodos*, un mecanismo que se explicará cuando se introduzca la programación orientada a objetos.

```python
temperaturas = [23.5, 24.0, 22.8, 19.0, 21.3]   # Lecturas horarias de un sensor
```

Como ya vimos, los elementos se acceden por su índice. El primero tiene índice `0`, y los índices negativos cuentan desde el final: `temperaturas[-1]` es siempre el último elemento, sin importar cuántos haya.

```python
temperaturas[0]    # Primera lectura del día
```

```python
temperaturas[-1]   # Lectura más reciente
```

La mutabilidad es lo que distingue a las listas de las tuplas. La forma más directa de aprovecharla es reemplazar un elemento existente: la misma notación de índice que ya conocemos, pero en el lado izquierdo de una asignación. Agregar o eliminar elementos también es posible, pero requiere métodos que veremos más adelante.

```python
temperaturas[2] = 23.1   # Corregimos una lectura errónea
temperaturas
```

Las funciones built-in que ya conocemos operan directamente sobre listas. `len()` devuelve cuántos elementos contiene, y `min()` y `max()` encuentran el menor y el mayor. Se aplican exactamente igual que sobre cualquier otra secuencia.

```python
len(temperaturas)   # Cuántas lecturas hay
```

```python
min(temperaturas)   # Temperatura mínima registrada
```

```python
max(temperaturas)   # Temperatura máxima registrada
```

La función `sum()` devuelve la suma de todos los elementos de una lista numérica. Combinada con `len()`, que ya conocemos, permite calcular el promedio de forma directa.

```python
suma = sum(temperaturas)
promedio = suma / len(temperaturas)
print(f"Temperatura promedio del día: {promedio:.2f} °C")  # :.2f indica que se muestran solo 2 decimales
```

La función `sorted()` recibe una secuencia y devuelve una *nueva* lista con los mismos elementos ordenados de menor a mayor, sin modificar la original. La distinción es importante: la lista original permanece intacta, y el resultado ordenado es un objeto nuevo.

```python
calificaciones = [7.5, 9.0, 8.5, 6.0, 10.0, 8.0]
print(sorted(calificaciones))   # sorted() devuelve una nueva lista; no modifica la original
print(calificaciones)           # La original no cambió
```

Las listas también pueden contener elementos de distintos tipos, incluidas otras listas. Esto permite representar estructuras más complejas, como una pequeña tabla de datos donde cada fila es a su vez una lista.

```python
# Cada fila contiene: nombre, edad, calificación
alumnos = [
    ["Ana",   20, 9.2],
    ["Luis",  22, 8.1],
    ["María", 21, 9.7]
]
```

```python
alumnos[0]       # Primera fila completa
```

```python
alumnos[0][2]    # Calificación de Ana: primer alumno, tercer campo
```

## Tuplas `tuple`

Una *tupla* es una colección ordenada e *inmutable*. Comparte con las listas la idea de posición y acceso por índice, pero su contenido no puede modificarse una vez creada.

Esta restricción no es un defecto: es una garantía. Hay datos que por su naturaleza no deberían cambiar. Las coordenadas geográficas de una ciudad son fijas. Los coeficientes de un modelo matemático no se modifican durante el cálculo. Las dimensiones de una imagen no deben alterarse una vez leída. Usar una tupla en esos casos comunica esa intención de forma explícita y protege los datos de modificaciones accidentales.

Las tuplas se definen con paréntesis `()`.

```python
cdmx = (19.4326, -99.1332)     # Coordenadas de Ciudad de México
gdl  = (20.6597, -103.3496)    # Coordenadas de Guadalajara
```

```python
cdmx[0]   # Latitud
```

```python
cdmx[1]   # Longitud
```

Intentar modificar un elemento de una tupla produce un `TypeError`. Python protege su contenido ante cualquier intento de reasignación, en coherencia con la garantía de inmutabilidad que la tupla ofrece.

```python
# Esta celda produce un error: las tuplas son inmutables
cdmx[0] = 0.0
```

Un uso muy natural de las tuplas es el *desempaquetado*. Para entenderlo, conviene partir de algo conocido: cuando escribimos `x = 10`, Python asocia el nombre `x` al valor `10`. El desempaquetado extiende esa idea a varios nombres a la vez: si el lado derecho es una tupla con $n$ elementos, podemos escribir $n$ nombres en el lado izquierdo separados por comas, y Python asignará cada elemento al nombre que le corresponde por posición.

La cantidad de nombres debe coincidir exactamente con la cantidad de elementos; de lo contrario, Python produce un error.

```python
latitud, longitud = cdmx   # desempaquetado: cada variable recibe un elemento de la tupla
print(f"Latitud:  {latitud}")
print(f"Longitud: {longitud}")
```

El desempaquetado también aparece de forma natural cuando se trabaja con listas de tuplas, donde cada elemento de la lista es un registro con varios campos.

```python
ciudades = [
    ("CDMX",        19.4326,  -99.1332),
    ("Guadalajara", 20.6597, -103.3496),
    ("Monterrey",   25.6866,  -100.3161)
]

nombre, lat, lon = ciudades[1]   # desempaquetado: los tres campos de la tupla van a tres variables
print(f"{nombre} está en ({lat}, {lon})")
```

El patrón es siempre el mismo: datos fijos, con campos de distintos tipos, que no deberían cambiar. Otros ejemplos naturales:

```python
rango_porcentaje = (0.0, 100.0)   # Valores mínimo y máximo permitidos
fecha_inicio     = (2025, 1, 15)  # Año, mes, día de inicio del estudio
resolucion       = (1920, 1080)   # Ancho y alto en píxeles
```

## Conjuntos `set`

Hasta ahora, todas las estructuras que hemos visto tienen algo en común: el orden importa. Las listas y las tuplas guardan sus elementos en una secuencia, y ese orden es parte de su identidad. Pero hay situaciones donde el orden es completamente irrelevante; lo que importa es únicamente *qué valores están presentes*.

Un *conjunto* es una colección no ordenada de elementos únicos. A diferencia de las listas y las tuplas, los conjuntos no mantienen ningún orden entre sus elementos y no permiten duplicados: si se intenta agregar un valor que ya existe, simplemente se ignora. Esta estructura modela directamente la idea matemática de conjunto, y resulta natural en problemas como: ¿qué materias distintas eligieron los alumnos de un grupo?, ¿qué opciones marcaron los encuestados?, ¿qué identificadores únicos aparecen en un registro de accesos? En todos esos casos, la multiplicidad es irrelevante y el orden no tiene ningún significado.

Los conjuntos se definen con llaves `{}`.

```python
# Respuestas a la pregunta: ¿cuál es tu materia favorita?
respuestas = {"Cálculo", "Estadística", "Álgebra", "Estadística", "Cálculo", "Física"}
respuestas   # Los duplicados se eliminan automáticamente
```

El resultado puede aparecer en cualquier orden: los conjuntos no garantizan ningún orden fijo. Esto tiene una consecuencia directa: no tiene sentido preguntar por el "primer" o el "segundo" elemento, así que el acceso por índice no está disponible. Lo que sí funciona es el operador `in`, que ya conocemos de las secuencias, y aquí se comporta exactamente igual.

```python
"Álgebra" in respuestas   # ¿Álgebra fue mencionada?
```

```python
"Historia" in respuestas  # ¿Historia fue mencionada?
```

La función `set()` puede recibir cualquier secuencia y convertirla en conjunto, eliminando automáticamente los duplicados. Esto hace que el patrón "obtener los valores únicos de una lista" sea inmediato.

```python
accesos = ["usr_3", "usr_1", "usr_3", "usr_2", "usr_1", "usr_3"]
set(accesos)   # set() convierte cualquier secuencia en conjunto, eliminando duplicados
```

### Operaciones entre conjuntos

Donde los conjuntos se vuelven especialmente expresivos es en las operaciones entre colecciones. Las mismas operaciones que se estudian en álgebra de conjuntos —unión, intersección, diferencia— están disponibles directamente en Python mediante operadores, y su semántica es exactamente la matemática.

La *unión* `|` produce un nuevo conjunto con todos los elementos que aparecen en cualquiera de los dos.

```python
grupo_manana = {"Ana", "Luis", "María", "Carlos"}
grupo_tarde  = {"María", "Pedro", "Carlos", "Sofía"}

grupo_manana | grupo_tarde   # Todos los estudiantes en cualquiera de los dos grupos
```

La *intersección* `&` produce únicamente los elementos presentes en ambos conjuntos al mismo tiempo.

```python
grupo_manana & grupo_tarde   # Estudiantes inscritos en ambos grupos
```

La *diferencia* `A - B` produce los elementos que están en `A` pero no en `B`.

```python
grupo_manana - grupo_tarde   # Estudiantes exclusivos del grupo de la mañana
```

Estas operaciones son directamente aplicables en análisis de datos. Por ejemplo, si un investigador tiene dos listas de genes identificados en experimentos distintos y quiere saber cuáles aparecieron en ambos, cuáles son exclusivos de cada experimento y cuántos genes distintos se encontraron en total, la respuesta se escribe de forma natural con conjuntos.

```python
experimento_1 = {"BRCA1", "TP53", "EGFR", "MYC"}
experimento_2 = {"TP53", "MYC", "KRAS", "PTEN"}

print("En ambos experimentos:", experimento_1 & experimento_2)
print("Solo en experimento 1:", experimento_1 - experimento_2)
print("Solo en experimento 2:", experimento_2 - experimento_1)
print("Total de genes únicos:", experimento_1 | experimento_2)
```

## Diccionarios `dict`

Hay una limitación que aparece con frecuencia al usar listas: los elementos se identifican por posición, y las posiciones son anónimas. Si representamos la información de un estudiante como `["Carlos", 21, "Actuaría", 8.7]`, tenemos que recordar que el nombre está en la posición `0`, la edad en la `1`, y así sucesivamente. Quien lea ese código sin contexto no tiene forma de saber qué significa cada número.

Un *diccionario* resuelve ese problema. Es una colección de pares *clave-valor*: en lugar de posiciones anónimas, cada dato tiene un nombre. Se accede a la información por clave, no por índice.

Los diccionarios se definen con llaves `{}`, usando la forma `clave: valor` para cada entrada, separadas por comas.

```python
estudiante = {
    "nombre":   "Carlos",
    "edad":     21,
    "carrera":  "Actuaría",
    "promedio": 8.7
}
```

Para recuperar el valor asociado a una clave, se usan corchetes con la clave entre comillas en lugar de un índice numérico. La mecánica es la misma que ya conocemos, pero ahora el "índice" es una palabra con significado.

```python
estudiante["nombre"]
```

```python
estudiante["promedio"]
```

Los diccionarios son mutables. Para modificar el valor de una clave existente, o para agregar una entrada nueva, se usa la misma sintaxis de asignación que ya conocemos: la clave entre corchetes en el lado izquierdo, el nuevo valor en el derecho.

```python
estudiante["promedio"] = 9.1   # Actualizamos el promedio
estudiante["semestre"] = 6     # Agregamos un campo nuevo
estudiante
```

El operador `in` también funciona con diccionarios, pero verifica si una *clave* está presente, no un valor.

```python
"carrera" in estudiante   # ¿Existe la clave 'carrera'?
```

```python
"beca" in estudiante      # ¿Existe la clave 'beca'?
```

Intentar acceder a una clave que no existe con la notación de corchetes produce un `KeyError`. La celda siguiente muestra ese error de forma controlada, con el acceso comentado para que no interrumpa la ejecución.

```python
# Acceso a una clave inexistente: produce un KeyError
# estudiante["beca"]   # error: la clave 'beca' no existe en el diccionario
```

### Más contextos de uso

Los diccionarios aparecen de forma natural en muchos problemas. Un patrón muy común es representar tablas de datos: en lugar de una lista de listas con posiciones arbitrarias, cada fila se representa como un diccionario donde las claves son los nombres de las columnas. El código resultante es mucho más legible.

```python
# Tabla de datos: cada elemento es un registro completo
grupo = [
    {"nombre": "Ana",   "edad": 20, "calificacion": 9.2},
    {"nombre": "Luis",  "edad": 22, "calificacion": 8.1},
    {"nombre": "María", "edad": 21, "calificacion": 9.7},
]
```

```python
grupo[0]                    # Registro completo del primer estudiante
```

```python
grupo[0]["calificacion"]    # Calificación del primer estudiante
```

Otro uso muy frecuente es el *conteo de frecuencias*: un diccionario donde las claves son los posibles valores de una variable y los valores son cuántas veces aparece cada uno. Por ejemplo, para contar cuántas veces marcaron cada opción los participantes de una encuesta.

```python
respuestas = ["A", "B", "A", "C", "B", "A", "D", "B", "A"]

frecuencias = {
    "A": 0,
    "B": 0,
    "C": 0,
    "D": 0
}

# Contamos manualmente cuántas veces aparece cada opción
for r in respuestas:              # for recorre cada elemento de la lista uno por uno
    frecuencias[r] = frecuencias[r] + 1   # incrementa el contador de la opción correspondiente

frecuencias
```

En la celda anterior aparece por primera vez un `for`, una estructura que aún no hemos presentado formalmente. Por ahora basta con entender que `for r in respuestas` recorre la lista un elemento por vez, asignando cada uno a `r`. La instrucción dentro del `for` actualiza el conteo correspondiente en el diccionario. Esta estructura —uno de los mecanismos más poderosos del lenguaje— se introduce con detalle en el siguiente notebook.

Con las cuatro estructuras que hemos visto —lista, tupla, conjunto, diccionario— ya disponemos de un vocabulario suficiente para organizar casi cualquier tipo de información. La pregunta que sigue naturalmente es cómo operar sobre esa información de forma sistemática: cómo recorrer una colección entera, cómo tomar decisiones en función de los datos, cómo automatizar procesos que de otra forma habría que escribir a mano. Eso es exactamente lo que abordan los notebooks que siguen.

```python
datos_clima = {
    "ciudad":      "Ciudad de México",
    "temperatura": 22.5,
    "humedad":     68,
    "condicion":   "Parcialmente nublado"
}

print(f"En {datos_clima['ciudad']} hay {datos_clima['temperatura']} °C.")
print(f"Condición: {datos_clima['condicion']}.")
```

## Resumen comparativo

Las cuatro estructuras que hemos visto responden a necesidades distintas. La siguiente tabla resume sus propiedades esenciales.

| Estructura | Sintaxis   | Ordenada | Mutable | Permite duplicados | Acceso        |
|:----------:|:----------:|:--------:|:-------:|:-----------------:|:-------------:|
| `list`     | `[a, b]`   | Sí       | Sí      | Sí                | Por índice    |
| `tuple`    | `(a, b)`   | Sí       | No      | Sí                | Por índice    |
| `set`      | `{a, b}`   | No       | Sí      | No                | Solo con `in` |
| `dict`     | `{k: v}`   | Sí*      | Sí      | No (en claves)    | Por clave     |

\* Los diccionarios conservan el orden de inserción a partir de Python 3.7.

Una regla práctica para elegir:

- Si los datos tienen un orden y pueden cambiar con el tiempo, usa una **lista**.
- Si los datos tienen un orden y no deben cambiar, usa una **tupla**.
- Si lo que importa es qué elementos están presentes, sin importar el orden ni la repetición, usa un **conjunto**.
- Si cada dato tiene un nombre o identificador propio, en lugar de una posición, usa un **diccionario**.

## Conversión entre estructuras

Python permite convertir entre estas estructuras usando las funciones `list()`, `tuple()` y `set()`. Estas conversiones son útiles cuando una colección se obtiene de cierta forma pero se necesita en otra.

Un patrón muy común es obtener los valores únicos de una lista convirtiéndola a conjunto, y luego regresar a lista para poder ordenarlos con `sorted()`.

```python
numeros = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]

unicos    = set(numeros)           # Elimina duplicados
ordenados = sorted(list(unicos))   # list() convierte el conjunto a lista para poder ordenarlo

print("Con duplicados: ", numeros)
print("Sin duplicados: ", ordenados)
```

También se puede convertir de lista a tupla cuando se tiene una colección que fue construida dinámicamente pero que a partir de cierto punto no debe modificarse: se construye como lista, se le agregan los elementos necesarios (operación que, como se mencionó, requiere métodos que veremos más adelante), y luego se convierte a tupla para protegerla.

```python
# Se construye dinámicamente la lista de coeficientes
coeficientes = [1.0, -3.5, 2.0]      # Representa 1x² - 3.5x + 2

# Una vez completa, la convertimos a tupla para que no pueda modificarse
coeficientes = tuple(coeficientes)   # tuple() convierte la lista en tupla inmutable
coeficientes
```
